## **SQL statement untuk menemukan jumlah buyer per country**

Menemukan jumlah buyer per country menggunakan SQL command COUNT dan GROUP BY

```python
sql4 = """ SELECT country, COUNT(buyerid) as total_buyer
FROM public.user_tab GROUP BY country;"""
buyer_per_country = pd.read_sql_query(sql4, conn)
buyer_per_country
```

## **SQL statement untuk menemukan jumlah total item_views untuk setiap shop**

Menemukan jumlah total item_views untuk setiap shop menggunakan SQL command **JOIN** antara table user dan performance, lalu **COUNT** total_views dan **GROUP BY** ditambah **ORDER BY** untuk di urutkan secara ascending.

```python
sql5 = """ SELECT t1.shopid, COUNT(t2.item_views) as total_views
FROM public.user_tab t1 INNER JOIN public.performance_tab t2
ON t1.shopid=t2.shopid GROUP BY t1.shopid ORDER BY shopid ASC;"""
item_views_shop = pd.read_sql_query(sql5, conn)
item_views_shop
```

## **SQL statement untuk menemukan berapa banyak jumlah order yang diterima oleh shop dalam interval 1 bulan**

Menggunakan **COUNT** dari orderid dari tabel order_tab di **JOIN** ke tabel user_tab untuk shopid. **GROUP BY** berdasarkan month. Untuk melihat perform shop per bulan.

```python
sql6 = """ SELECT shopid, COUNT(order_id) as total_order, DATE_TRUNC ('month', order_time) as order_month
FROM public.order_tab
GROUP BY shopid, DATE_TRUNC ('month', order_time) ORDER BY total_order DESC, order_month, shopid ASC;"""
jumlah_order_bulanan = pd.read_sql_query(sql6, conn)
jumlah_order_bulanan
```

## **SQL statement untuk menemukan TOP 10 buyer by GMV dari setiap country**

Menggunakan **COUNT** GMV dari table order lalu **JOIN** untuk table order dan user by buyerid, **ORDER BY DESC** lalu di **LIMIT** sebanyak 10

```python
sql7 = """
SELECT t1.buyerid, COUNT(t1.gmv) as gmv, t2.country
FROM public.order_tab t1
INNER JOIN public.user_tab t2 ON t1.buyerid=t2.buyerid
GROUP BY t1.buyerid, t2.country
ORDER BY gmv DESC
LIMIT 10
;"""
top_10_buyer = pd.read_sql_query(sql7, conn)
top_10_buyer
```

## **SQL statement untuk latest order date dari setiap buyer**

Menggunakan table order yang di **GROUP BY** order_time dan buyer id lalu **ORDER BY DESC** untuk order_time dan **ASC** untuk buyerid

```python
sql8 = """
SELECT buyerid, order_time
FROM order_tab
GROUP BY order_time, buyerid
ORDER BY order_time DESC, buyerid ASC
;"""
latest_order_buyer = pd.read_sql_query(sql8, conn)
latest_order_buyer
```
